### Imputing Missing Values for Categorical variables using Mode

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values = np.nan,
                        strategy ='most_frequent')

# Fitting the data to the imputer object
imputer = imputer.fit(Initial_ExoPlt[car_var])

# Imputing the data
data_cat = imputer.transform(Initial_ExoPlt[car_var])
```

```python
data_cat = pd.DataFrame(data_cat,columns=Initial_ExoPlt[car_var].columns)
```

```python
data_cat.info()
```

### Imputing Missing Values for Numerical variables using SimpleImputer

```python
imputer_num = SimpleImputer(missing_values = np.nan,
                        strategy ='mean')

# Fitting the data to the imputer object
imputer_num = imputer.fit(Initial_ExoPlt[num_var])

# Imputing the data
data_num = imputer.transform(Initial_ExoPlt[num_var])

data_num = pd.DataFrame(data_num,columns=Initial_ExoPlt[num_var].columns)
```

### One hot encoding the categorical variables

```python
car_var
car_var_encoded_set2 = ['koi_disposition','koi_pdisposition','koi_tce_delivname']
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_columns = data_cat[car_var_encoded_set2].columns.tolist()

encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(data_cat[car_var_encoded_set2])
one_hot_df = pd.DataFrame(one_hot_encoded,columns=encoder.get_feature_names_out(categorical_columns))
```

```python
encoder.get_feature_names_out(categorical_columns)
```

```python
df_encoded = pd.concat([data_cat, one_hot_df], axis=1)

# Drop the original categorical columns
df_encoded = df_encoded.drop(car_var_encoded_set2, axis=1)
```

```python
df_encoded.info()
```

```python
for var in df_encoded.columns:
    df_encoded[var] = df_encoded[var].astype("category")
```
